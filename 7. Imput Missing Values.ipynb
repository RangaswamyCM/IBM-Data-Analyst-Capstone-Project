### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
duplictes=df[df.duplicated()]
print("Duplicate rows")
duplictes
#print("Number of duplicate rows:", df.duplicated().sum())
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here
df.drop_duplicates(inplace=True)
df
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
missing_Values=df.isnull().sum()
missing_Percetage=(df.isnull().sum()/len(df))*100
missing_report=pd.DataFrame({"Missing Values":missing_Values, "Missing_Values(%)":missing_Percetage})
missing_report
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here
Missing_Remotework=df['RemoteWork'].isnull().sum()
print("Missing values in column 'RemoteWork':",Missing_Remotework)
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here
value_count=df["RemoteWork"].value_counts(dropna=False)
print("Value Count for Each Unique value in column RemoteWork':",value_count)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
most_frqnt_Value=df["RemoteWork"].mode()[0]
print("Most frquent Value in column 'RemoteWork':",most_frqnt_Value)
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
df["RemoteWork"]=df["RemoteWork"].fillna(most_frqnt_Value)
print("missing values after filled the missing values with most repited value:",df["RemoteWork"].isnull().sum())
```

##### Task 8: Check for any compensation-related columns and describe their distribution.
