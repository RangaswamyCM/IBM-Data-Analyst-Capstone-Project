### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# Get unique values in the 'Country' column
unique_countries = df['Country'].unique()
print(unique_countries)
```

```python
unique_countries = df['EdLevel'].unique()
print(unique_countries)
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
country_mapping = {
    # Original inconsistent / unusual entries
    'Republic of Korea': 'South Korea',
    "Democratic People's Republic of Korea": 'North Korea',
    'United Republic of Tanzania': 'Tanzania',
    'Venezuela, Bolivarian Republic of...': 'Venezuela',
    'Iran, Islamic Republic of...': 'Iran',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'Nomadic': None,  # mark as missing
    
    # Additional entries to standardize
    "Côte d'Ivoire": "Ivory Coast",
    "Russian Federation": "Russia",
    "Hong Kong (S.A.R.)": "Hong Kong",
    "Micronesia, Federated States of...": "Micronesia",
    "Brunei Darussalam": "Brunei",
    "Libyan Arab Jamahiriya": "Libya",
    "Congo, Republic of the...": "Congo",
    "Lao People's Democratic Republic": "Laos",
    "Bolivarian Republic of Venezuela...": "Venezuela"
}
df['Country'] = df['Country'].replace(country_mapping)
df['Country']
```

```python
# Step 3: Normalize EdLevel values

# 1. Create a mapping dictionary for clarity (optional but good practice)
edlevel_mapping = {
    'Primary/elementary school': 'Primary/Elementary School',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary School',
    'Some college/university study without earning a degree': 'Some College/University (No Degree)',
    'Associate degree (A.A., A.S., etc.)': 'Associate Degree',
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': "Bachelor's Degree",
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': "Master's Degree",
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional Degree (PhD/MD/JD/etc.)',
    'Something else': 'Other',
}

# 2. Apply mapping
df['EdLevel_Cleaned'] = df['EdLevel'].map(edlevel_mapping)

# 3. Fill NaN with 'Unknown' for completeness
df['EdLevel_Cleaned'] = df['EdLevel_Cleaned'].fillna('Unknown')

# 4. Verify results
print(df['EdLevel_Cleaned'].value_counts(dropna=False))
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
print(df['Employment'].unique())


# Step 1: Split multiple employment types into separate entries
df['Employment'] = df['Employment'].fillna('I prefer not to say')  # handle missing
employment_dummies = df['Employment'].str.get_dummies(sep=';')

# Step 2: Clean column names (strip spaces)
employment_dummies.columns = employment_dummies.columns.str.strip()

# Step 3: Merge with main dataframe
df = pd.concat([df, employment_dummies], axis=1)

# Step 4: Display the new one-hot encoded columns
print(employment_dummies.head())
print("\nNumber of new columns created:", len(employment_dummies.columns))
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_values=df.isnull().sum().sort_values(ascending=False)
missing_values
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
mean_value = df['ConvertedCompYearly'].mean()
df.fillna({'ConvertedCompYearly':mean_value})
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
mode_value = df['RemoteWork'].mode()[0]
print("Most frequent value:", mode_value)
df.fillna({'RemoteWork':mode_value})
```
